```python
#joins staff and store tables if the store_id's match up 
```

```python
!psql -c "SELECT * FROM pagila.staff INNER JOIN pagila.store ON pagila.staff.store_id = pagila.store.store_id"
```

```python
#joins film and actor tables if the film_id's and the last updated times match up 
```

```python
!psql -c "SELECT * FROM pagila.film INNER JOIN pagila.film_actor ON pagila.film.film_id = pagila.film_actor.film_id AND pagila.film.last_update = pagila.film_actor.last_update"
```

```python
#joins address and citytables if the city_id's match up 
```

```python
!psql -c "SELECT * FROM pagila.address INNER JOIN pagila.city ON pagila.address.city_id = pagila.city.city_id LIMIT 10"
```

```python
#Joins address with city based on whether the city id's match up, match based on left column
```

```python
!psql -c "SELECT * FROM pagila.address LEFT JOIN pagila.city ON pagila.address.city_id = pagila.city.city_id LIMIT 10"
```

```python
#Joins address with customer based on whether the address_id's match up, match based on left column
```

```python
!psql -c "SELECT * FROM pagila.address LEFT JOIN pagila.customer ON pagila.address.address_id = pagila.customer.address_id LIMIT 10"
```

```python
#Joins address with city based on whether the city id's match up, match based on right column
```

```python
!psql -c "SELECT * FROM pagila.address RIGHT JOIN pagila.city ON pagila.address.city_id = pagila.city.city_id LIMIT 10"
```

```python
#Joins film table to itself based on the movies that have the same release year, making sure there are no repeats
```

```python
!psql -c "SELECT T1.title, T2.title, T2.release_year FROM pagila.film T1, pagila.film T2 WHERE T1.film_id != T2.film_id AND T1.release_year=T2.release_year LIMIT 10"
```
