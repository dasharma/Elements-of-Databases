```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

df = pd.read_csv('fs_shopify/reviews.csv', sep=',', header=0, lineterminator='\n')
reviews_rows = df.values.tolist()


for reviews_row in reviews_rows:
    reviews_record = {}
    reviews_record['app_id'] = row[0]
    reviews_record['author'] = row[1]
    reviews_record['rating'] = row[2]
    reviews_record['posted_at'] = row[3]
    
    reviews_ref = db.collection('reviews').document(reviews_row[0])
    
    batch.set(reviews_ref, reviews_record)
    
    
    batch.commit()
```

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

df = pd.read_csv('fs_shopify/categories.csv', sep=',', header=0, lineterminator='\n')
category_rows = df.values.tolist()
df = pd.read_csv('fs_shopify/apps.csv', sep=',', header=0, lineterminator='\n')
rows = df.values.tolist()

for category_row in category_rows:
    category_record = {}
    category_record['category_id'] = row[0]
    category_record['title'] = row[1]
    category_record['Apps'] = row[2]
    
    category_ref = db.collection('category').document(category_row[0])
    
    batch.set(category_ref, category_record)


    for row in rows: 
        if category_row[0] == row[0]:
            record = {}
            record['id'] = row[0]
            record['url'] = row[1]
            record['title'] = row[2]
            record['developer'] = row[3]
            record['developer_link'] = row[4]
            record['icon'] = row[5]
            record['rating'] = row[6]
            record['reviews_count'] = row[7]
            apps_ref = db.collection('apps').document(row[0])
            batch.set(apps_ref, record)
    
    
    batch.commit()
```

```python
from google.cloud.firestore_v1.base_query import FieldFilter
from google.cloud.firestore_v1 import aggregation

db = firestore.Client()
apps_ref = (db.collection('apps')).where(filter=FieldFilter('id', '>', ''))
aggregate_query = aggregation.AggregationQuery(apps_ref)
aggregate_query.count(alias="all")
results = aggregate_query.get()
for result in results:
    print (f"Number of documents: {result[0].value}")
```

```python
from google.cloud import firestore

db = firestore.Client()
query = db.collection('category').collection('apps').where('title', '==', 'Productivity')
    .order_by('rating', direction=firestore.Query.DESCENDING)\
    .limit(10)
results = query.stream()

for result in results:
    app_data = result.to_dict()
    app_id = result.id
    title = app_data['title']
    developer = app_data['developer']
    rating = app_data['rating']
    reviews_count = app_data['reviews_count']
    
    print(f'App ID: {app_id}')
    print(f'Title: {title}')
    print(f'Developer: {developer}')
    print(f'Rating: {rating}')
    print(f'Reviews Count: {reviews_count}')
```
